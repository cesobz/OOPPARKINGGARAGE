```python
class Car:
    
    def __init__(self, licensePlate, cardNumber, cardEXP):
        self.licensePlate = licensePlate
        self.cardNumber = cardNumber
        self.cardEXP = cardEXP
        
    def __repr__(self):
        return f'{self.licensePlate}, {self.cardNumber}, {self.cardEXP}'

class parkingGarage:

    def __init__(self):
        self.cars_added = []
        self.parkingSpaces = 10
        self.currentTicket = {}
        self.bill = 0

    def availability(self):
        return self.parkingSpaces

    def takeTicket(self, car):

        self.identifier = ['A1', 'B1', 'C1', 'D1', 'E1', 'F1', 'G1', 'H1', 'I1', 'J1',
                        'K1', 'L1', 'M1', 'N1', 'O1', 'P1', 'Q1', 'R1', 'S1', 'T1']

        if self.parkingSpaces > 0:

            self.cars_added.append(str(car).split(', '))

            self.parkingSpaces -= 1

            self.currentTicket = {'Info': [], 'license plate': [], 'cardNumber': [], 'cardEXP': []}

            for index, i in enumerate(self.cars_added):

                self.currentTicket['Info'].append(self.identifier[index])
                self.currentTicket['license plate'].append(i[0])
                self.currentTicket['cardNumber'].append(i[1])
                self.currentTicket['cardEXP'].append(i[2])
            return "car successfully added to the parking lot"

        else:
            print(f"Currently {self.parkingSpaces} parkingSpaces are available. We Apologize.")
    def leaveGarage(self, ticketInfo, payForParking):
        past_len = len(self.currentTicket['Info'])

        if ticketInfo not in self.currentTicket['Info']:
            print("We could not find your car. Are you sure you parked your car here? ")

        else:
            for index, value in enumerate(self.currentTicket['Info']):
                if value == ticketInfo:

                    print("Your car's license plate is:", self.currentTicket['license plate'][index])
                    print("Your cardNumber is:", self.currentTicket['cardNumber'][index])
                    print("Your cardEXP is :", self.currentTicket['cardEXP'][index])
                    
                    removed_car_index = self.currentTicket['Info'].pop(index)
                    self.currentTicket['license plate'].pop(index)
                    self.currentTicket['cardNumber'].pop(index)
                    self.currentTicket['cardEXP'].pop(index)
                    self.parkingSpaces += 1

        if len(self.currentTicket['Info']) < past_len:
            while True:
                if payForParking.isnumeric():
                    list_of_time_and_Info = [payForParking, removed_car_index]
                    break

                else:
                    print("Your input must be an integer. Sample input: 5 ")

                payForParking = input("Enter how long you were on the parking lot in hours or 'q' to quit. Example input: 5  -->  ")
                if payForParking in ['q', 'Q']:
                    break

            if int(list_of_time_and_Info[0]) < 10:
                self.bill = int(list_of_time_and_Info[0]) * 5
                return f'${self.bill} will be billed to card. Thank You, have a nice day'
            
            else:
                self.bill = int(list_of_time_and_Info[0]) * 5 + 100
                return f' ${self.bill} will be billed to card. Thank You, have a nice day'

    def inventory(self):
            for i in self.currentTicket.items():
                print(i)
                
myGarage = parkingGarage()
print(myGarage.availability())
myGarage.takeTicket(Car('NIBU2K', '4000-6872-6845-4956', '08/27'))
myGarage.takeTicket(Car('5Y4T1T', '4200-4589-1674-2841', '02/25'))
myGarage.takeTicket(Car('PB4UGO', '4400-6548-8498-3158', '10/23'))
myGarage.takeTicket(Car('E0SIK5', '4400-2197-6664-9997', '11/25'))
myGarage.inventory()
print(myGarage.leaveGarage('B1', '5'))
print(myGarage.leaveGarage('D1', '10'))
print(myGarage.availability())
                
```
